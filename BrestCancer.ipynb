```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, roc_curve, auc
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import joblib

df = pd.read_csv('breast cancer.csv')

X = df.drop(columns=['diagnosis'])
y = df['diagnosis']

imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

def train_evaluate_plot_roc(X_train, y_train, X_test, y_test):
    models = {
        'Logistic Regression': LogisticRegression(random_state=0),
        'Decision Tree': DecisionTreeClassifier(random_state=0, criterion='entropy'),
        'Random Forest': RandomForestClassifier(random_state=0, criterion='entropy', n_estimators=10),
        'Gradient Boosting': GradientBoostingClassifier(random_state=0),
        'Support Vector Machine': SVC(probability=True, random_state=0),
        'K-Nearest Neighbors': KNeighborsClassifier()
    }
    
    best_model = None
    best_accuracy = 0
    best_model_name = ""

    plt.figure(figsize=(10, 8))

    for name, model in models.items():
        model.fit(X_train, y_train)

        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

        train_accuracy = accuracy_score(y_train, y_train_pred)
        test_accuracy = accuracy_score(y_test, y_test_pred)

        print(f'{name} - Training Accuracy: {train_accuracy:.4f}')
        print(f'{name} - Testing Accuracy: {test_accuracy:.4f}')
        print('-' * 40)

        if test_accuracy > best_accuracy:
            best_accuracy = test_accuracy
            best_model = model
            best_model_name = name


        y_test_prob = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_test_prob)
        roc_auc = auc(fpr, tpr)

        plt.plot(fpr, tpr, lw=2, label=f'{name} (AUC = {roc_auc:.2f})')

    print(f'Best Model: {best_model_name} with Testing Accuracy: {best_accuracy:.4f}')
    joblib.dump(best_model, 'breastCancerModel.joblib')
    print(f'{best_model_name} model saved as best_model.joblib')

    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()

train_evaluate_plot_roc(X_train_scaled, y_train_encoded, X_test_scaled, y_test_encoded)
```

```python
# Load the saved model
loaded_model = joblib.load('breastCancerModel.joblib')

# Assuming new_data is a DataFrame containing the new data to be tested
# Replace this with your actual new data
new_data = pd.read_csv('data.csv')

# Preprocess the new data
X_new = new_data.drop(columns=['diagnosis'])
y_new = new_data['diagnosis']

# Handle missing values
X_new_imputed = imputer.transform(X_new)

# Standardize the features
X_new_scaled = scaler.transform(X_new_imputed)

# Encode the target variable
y_new_encoded = le.transform(y_new)

# Make predictions on the new data
y_new_pred = loaded_model.predict(X_new_scaled)

# Evaluate the model's performance on the new data
new_accuracy = accuracy_score(y_new_encoded, y_new_pred)
print(f'Accuracy on new data: {new_accuracy:.4f}')

# Compute ROC curve and ROC area for the new data
y_new_prob = loaded_model.predict_proba(X_new_scaled)[:, 1]
fpr_new, tpr_new, _ = roc_curve(y_new_encoded, y_new_prob)
roc_auc_new = auc(fpr_new, tpr_new)

# Plot the ROC curve for the new data
plt.figure(figsize=(10, 8))
plt.plot(fpr_new, tpr_new, lw=2, label=f'New Data (AUC = {roc_auc_new:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for New Data')
plt.legend(loc="lower right")
plt.show()
```
